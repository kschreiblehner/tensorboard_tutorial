# TensorBoard - Installation und Starten
Dieses Jupyter-Notebook ist Teil des Trainings 'Konfiguration und Nutzung von TensorBoard'. Es dient als Hilfestellung zur Installation und zum Starten von Tensorboard.

Betrachtete Themen ist die Installation mittels
- pip,
- Jupyter-Notebook und
- Docker.

## Abhängigkeiten

### Lokale Installation

Sofern TensorBoard __lokal__ genutzt werden soll, gelten die nachstehenden Abhängigkeiten. Eine Python 3.x Installation bildet die Grundlage. Diese kann auf <a href='https://www.python.org/'>python.org</a> heruntergeladen und installiert werden. Ferner ist eine Installation von TensorFlow notwendig. Dies kann innerhalb der Python-Umgebung mittels `pip install tensorflow` ausgeführt werden. Ferner kann man TensorFlow über <a href='https://www.tensorflow.org/'>tensorflow.org</a> beziehen.

### Docker-Installation

Bei der Installation mittels __Docker__ ist lediglich eine Docker-Installtion notwendig. Diese kann über <a href='https://www.docker.com/'>docker.com</a> bezogen werden.

## Installation von TensorBoard

### Lokale Installation 

- Mittels __pip__: Die Installation von TensorBoard kann mittels `pip install tensorboard` innerhalb der Python-Umgebung in einer Konsole durchgeführt werden.
- Innerhalb eines __Jupyter-Notebook__: Die Installation innerhalb eines Jupyter-Notebooks kann durch ausführen des Befehls `!pip install tensorboard` innerhalb einer Code-Zelle das entsprechende Paket in die Python-Umgebung installiert werden.

### Docker-Installation

Zur Installation von TensorBoard mittels Docker ist ein TensorFlow-Image notwendig. Dieses kann über den <a href='https://hub.docker.com/r/tensorflow/tensorflow/'>Tensorflow-Docker-Hub</a> erhalten werden. Innerhalb einer Konsole kann über den Befehl `docker pull tensorflow/tensorflow:latest` ein entsprechendes Docker-Image heruntergeladen werden.

## Starten von TensorBoard

TensorBoard nutzt Log-Dateien zur Auswertung. Dazu sind die entsprechenden Zugriffrechte auf einen Log-Ordner notwendig. Alle Option zum Starten sind über die Hilfe von TensorBoard ersichtlich. Diese kann mittels anhängen der Option `-h` beziehungsweise `--help` an den TensorBoard Befehl innerhalb eines Jupyter-Notebook `%tensorboard` oder der Konsole `tensorboard` aufgerufen werden. Die Wichtigsten sind 
- `--logdir PATH`: Pfad des Ordners der Log-Dateien,
- `--host ADDRESS`: Adresse, unter der TensorBoard verwendet werden soll (Standard: localhost) und
- `--port PORT`: Der Port, über den mit TensorBoard kommuniziert werden soll (Standard: 6006).

### Innerhalb eines Jupyter-Notebook

TensorBoard kann innerhalb eines Jupyter-Notebook interaktiv genutzt werden. Dafür muss die TensorBoard-Erweiterung mittels 

`%load_ext tensorboard` 

geladen werden. Anschließend kann TensorBoard mit Hilfe des Befehls

`%tensorboard --logdir PATH`

der Webservice für Tensorboard gestartet werden.

### Über die Browseroberfläche

Zum Starten kann in der Konsole nach Festlegung der entsprechenden Python-Umgebung der Befehl 

`tensorboard --logdir PATH` 

ausgeführt werden. 

Falls weder der Host, noch der Port spezifiziert wurden, ist die Weboberfläche im Browser unter 

`localhost:6006` 

zu sehen.


### Über Docker

Das Docker-Image kann über den Befehl 

<code>
docker run -d -p HOST_PORT:TB_PORT \
        --mount type=bind,source=LOG_DIR,target=/tensorboard_log \
    tensorflow/tensorflow:latest \
    tensorboard \
        --logdir /tensorboard_log --host ADDRESS \
        --port TB_PORT
</code> 

gestartet werden. Die Option `-d` startet den Docker-Containers im Hintergrund. Die Option `-p` spezifiziert, über welchen Port auf dem Host-System (`HOST_PORT`) auf TensorBoard innerhalb des Containers zugegriffen werden soll. Der Wert `TB_PORT` gibt an, auf welchen Port TensorBoard innerhalb des Containers Daten sendet. `--mount` bindet das Verzeichnis `LOG_DIR` des Host-Systems im Container in das Verzeichnis `/tensorboard_log` ein. Am Ende startet der Befehl `tensorboard` den Webservice TensorBoard innerhalb des Containers. Dort liest Tensorboard die Log-Files aus dem Ordner `/tensorboard_log` aus. Ferner kommuniziert TensorBoard innerhalb des Docker-Containers mit dem Port `TB_PORT`. Auf dem Host-System können anschließend alle Log-Dateien, die über TensorBoard ausgewertet werden sollen, in das Verzeichnis `LOG_DIR` gespeichert werden. `ADDRESS` gibt dabei die Adresse des Host-Systems an. Statt `localhost` sollte dort im Allgemeinen `0.0.0.0` verwendet werden. Siehe dazu <a href='https://nickjanetakis.com/blog/docker-tip-54-fixing-connection-reset-by-peer-or-similar-errors'>diesen Artikel</a>.

Ein Beispiel zum Starten ist:
<code>
docker run -d -p 6100:6100 \
         --mount type=bind,source=/tmp/tensorboard_log,target=/tensorboard_log \
     tensorflow/tensorflow:latest \
     tensorboard \
         --logdir /tensorboard_log --host 0.0.0.0 \
         --port 6100
</code>
